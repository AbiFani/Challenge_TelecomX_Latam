```python
# guardar en un dataframe la columna customerID
clientes_customerID = datos['customerID']
clientes_customerID
```

```python
# guardar en un dataframe la columna Churn
clientes_Churn = datos['Churn']
clientes_Churn
```

```python
# Normalizando archivo Json columna customer
clientes_normalizados = pd.json_normalize(datos['customer'])
clientes_normalizados




```

```python
# Nomalizando columna phone
cliente_telefono = pd.json_normalize(datos['phone'])
cliente_telefono
```

```python
# Nomalizando columna internet
cliente_internet = pd.json_normalize(datos['internet'])
cliente_internet
```

```python
# Nomalizando columna account
cliente_account = pd.json_normalize(datos['account'])
cliente_account
```

```python
# guadar todo en un nuevo dataframe llamado datos_nomalizados, las columnas normalizadas

datos_normalizados = pd.concat([clientes_customerID,clientes_Churn, clientes_normalizados, cliente_telefono, cliente_internet, cliente_account], axis=1)
```

```python
# mostrar dataframe
datos_normalizados
```

```python
# dataframe con el nombre de las columnas
columnas = list(datos_normalizados.columns)
columnas
```

```python
# explorar con el metodo explode, para tener datos individualmente
datos_normalizados = datos_normalizados.explode(columnas)
# datos_normalizados
```

```python
# mostrar primeras filas
datos_normalizados.head()
```

```python
# reemplazar datos en las columnas del dataframe
datos_normalizados = datos_normalizados.replace(' ', np.nan)
datos_normalizados
```

```python
# Transformar columna Charges.Total a tipo de dato decimal , con float64
datos_normalizados['Charges.Total'] = datos_normalizados['Charges.Total'].astype(np.float64)
```

```python
# mostrar el cambio en la columna
datos_normalizados.info()
```

```python
# mostrar el cambio tipo de datos
datos_normalizados.dtypes
```
